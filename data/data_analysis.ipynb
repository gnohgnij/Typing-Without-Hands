```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(eye_coor, gaze_coor)
```

```python
model.score(eye_coor, gaze_coor)
```

```python
plt.rcParams.update({'font.size': 40})
plt.rcParams["figure.figsize"] = (20,10)

actual_x = plt.scatter([x[0] for x in eye_coor], [x[0] for x in gaze_coor], s=10**2, marker='o', c='red')
actual_y = plt.scatter([x[1] for x in eye_coor], [x[1] for x in gaze_coor], s=10**2, marker='x', c='red')

predict = model.predict(eye_coor)

predict_x = plt.scatter([x[0] for x in eye_coor], [x[0] for x in predict], s=10**2, marker='o', c='blue')
predict_y = plt.scatter([x[1] for x in eye_coor], [x[1] for x in predict], s=10**2, marker='x', c='blue')

plt.legend((actual_x, actual_y, predict_x, predict_y),
           ('tracking point x vs actual gaze point x', 'tracking point y vs actual gaze point y', 'tracking point x vs predicted gaze point x', 'tracking point y vs predicted gaze point y'),
           scatterpoints=1,
           loc='upper left',
           ncol=2,
           fontsize=20)

plt.show()
```
